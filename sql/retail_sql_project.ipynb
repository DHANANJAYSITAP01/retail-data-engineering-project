**Data Quality & Schema Validation**

```python
%sql
SELECT * FROM customers;
SELECT * FROM orders;
SELECT * FROM products;
SELECT * FROM payments;
SELECT * FROM returns;
```

**Null check**

```python
%sql
select * from customers where customer_id is null;
select * from orders where sales is null;
```

**Data Consistency & Integrity Checks**

```python
%sql
select distinct customer_id from orders;
select distinct product_id from products;
```

**Customer Order Transaction Mapping (Join Layer)**

```python
%sql
SELECT 
    c.customer_name,
    o.order_id,
    o.sales,
    p.product_name
FROM customers c
JOIN orders o 
ON c.customer_id = o.customer_id
JOIN products p 
ON o.product_id = p.product_id;
```

**Customer-wise Revenue Analysis**

```python
%sql
SELECT 
    c.customer_name,
    SUM(o.sales) AS total_revenue
FROM customers c
JOIN orders o 
ON c.customer_id = o.customer_id
GROUP BY c.customer_name
ORDER BY total_revenue DESC;
```

**product-wise Sales Performance**

```python
%sql
select p.product_name,sum(o.sales)as total_sales from products p inner join orders o on p.product_id=o.product_id group by p.product_name order by total_sales ;
```

**Payment Status Distribution Analysis**

```python
%sql
select payment_status ,count(*) from payments group by payment_status; 
```

**Return Rate Analysis (Customer Behavior)**

```python
%sql
select return_reason, count(*) from returns group by return_reason;
```

**Top 5 High Value Customers**

```python
%sql
SELECT 
    c.customer_name,
    SUM(o.sales) AS total_spent
FROM customers c
JOIN orders o 
ON c.customer_id = o.customer_id
GROUP BY c.customer_name
ORDER BY total_spent DESC
LIMIT 5;
```

**End-to-End Customer Business View**

```python
%sql
Select c.customer_name,sum(o.sales)as total_spend,count(o.order_id)as total_orders from customers c inner join orders o on c.customer_id=o.customer_id group by c.customer_name ;
```

**Top Selling Product Categories**

```python
%sql
select p.category,sum(o.sales) as total_sales from orders o inner join products p on o.product_id=p.product_id group by p.category order by total_sales desc;
```

**Monthly Sales Trend Analysis**

```python
%sql
select month(order_date) as month,sum(sales) as total_sales from orders group by month order by total_sales desc;
```
